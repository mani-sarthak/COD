```python
import pandas as pd

# Load the CSV files into pandas DataFrames
built_matched_df = pd.read_csv('./../built_matched.csv')
control_points_df = pd.read_csv('control_points.csv')

# Merge the DataFrames based on the 'UID_matched' column
merged_df = pd.merge(built_matched_df, control_points_df[['UID', 'ndvi_Kharif_2016', 'ndvi_Kharif_2018',
                                                           'ndvi_Kharif_2019', 'ndvi_Kharif_2020', 'ndvi_Kharif_2021',
                                                           'ndvi_Rabi_2016', 'ndvi_Rabi_2018', 'ndvi_Rabi_2019',
                                                           'ndvi_Rabi_2020', 'ndvi_Rabi_2021', 'ndvi_Zaid_2016',
                                                           'ndvi_Zaid_2018', 'ndvi_Zaid_2019', 'ndvi_Zaid_2020',
                                                           'ndvi_Zaid_2021']], left_on='UID_matched', right_on='UID', how='left')
# Drop redundant columns
merged_df = merged_df.drop(['UID_y'], axis=1)

# Rename the UID_x column to UID
merged_df = merged_df.rename(columns={'UID_x': 'UID'})



# Save the result to a new CSV file
merged_df.to_csv('built_matched_NDVI.csv', index=False)
```
