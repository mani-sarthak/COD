```python
import os

def get_directories(path):
    # Use list comprehension to filter only directories
    directories = [d for d in os.listdir(path) if os.path.isdir(os.path.join(path, d))]
    return directories

# Replace 'your_directory_path' with the path of the directory you want to inspect
directory_path = './addStates/2023/'

directories_list = get_directories(directory_path)

print("Directories in {}: {}".format(directory_path, directories_list))

```

```python
districtStateMap = dict()
for state in directories_list:
    districts = get_directories(directory_path + state)
    for district in districts:
        districtStateMap[district] = state
        

    
```
